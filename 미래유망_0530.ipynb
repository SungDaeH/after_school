# 머신러닝의 순서

1. 데이터 전처리
2. 데이터 세트 분리
3. 모델 학습 및 검증 평가
4. 예측 수행

# 데이터 전처리의 종류

- 데이터 클리닝(cleaning) 
```
오류가 있거나, 손상되었거나, 형식에 문제가 있거나, 중복되었거나, 어떤 형태로든 불완전한 데이터를 수정하거나 제거하는 프로세스이다.
```
- 데이터 통합(integration)
```
데이터에서 더 높은 가치를 창출하고 유용한 정보를 얻기 위해 이러한 개별 데이터의 수집을 통합하는 프로세스이다.
```
- 데이터 변환(transformation)
```
데이터 소스에서 추출된 가공되지 않은 데이터를 사용 가능한 데이터세트로 변환하는 프로세스이다.
```
- 데이터 축소(redution)
```
원래 용량 기준보다 작은 양의 데이터 표현결과를 얻게 되더라도 원 데이터의 완결성을 유지하기 위해 사용된다.
```
- 데이터 이산화(discretization)
```
연속형 속성의 값을 구간으로 나누어 범주형 속성으로 변환하여 데이터의 크기를 줄인다.
```

# 인코딩과 디코딩

- 인코딩(encoding) : 데이터 압축, 다른 형식으로 변환하여 저장 공간 절약, 전송 시간을 줄이는 데 도움이 된다.

- 디코딩(decoding) : 인코딩된 데이터를 원래의 형태로 되돌려 사용자가 이해할 수 있게 만드는 역할을 한다.
